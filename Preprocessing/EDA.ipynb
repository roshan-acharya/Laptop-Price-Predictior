```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('../Data/Cleaned/mudita_laptops_cleaned.csv')
```

```python
df.info()
```

```python
df.head(5)
```

```python
#PLot for price vs ram
df.groupby("Ram").Price.mean().plot(kind='bar', figsize=(10, 5), color='blue')
```

```python
df['Processor'].unique()
```

```python
sns.histplot(df['Price'], bins=20, kde=True, color='blue')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#correlation between RAM,Storage, and Price
correlation_matrix = df[['Price', 'Ram', 'Storage']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

```python
sns.lineplot(data=df, x='Ram', y='Price', marker='o')
```

```python
sns.pairplot(df[['Price', 'Ram', 'Storage']], diag_kind='kde', markers='o')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Storage', y='Price', data=df)
```
